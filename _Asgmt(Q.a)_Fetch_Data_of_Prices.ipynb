# a. Write a python script to fetch data of prices for the year 2020 (date wise from 1stJan’2020 to 31st Dec’2020) for district “Agra” of Uttar Pradesh from the data sources.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
```

```python
# Lets read the datset... Here dataset is in the html form that's y pd.rea_html()is used..

dt = pd.read_html('Agmarknet_Price_Report.xls')
dt
```

```python
# Convert data to dataframe.

data = dt[0]
```

```python
#Lets check if there is any missing values present in the dataset

data.isnull().sum()
```

```python
# Now we will convert Price Date to datetime object.

data["Price Date"]=pd.to_datetime(data["Price Date"])
print()
print()
print(" Data of prices for the year 2020 (date wise from 1st Jan’2020 to 31st Dec’2020) for district “Agra” of Uttar Pradesh")
print("---------------------------------------------------------------------------------------------------------------------")


# To get records date wise from Jan 2020 to Dec 2020 arrange "Price Date" in ascending order.

pricedate = data[data["Price Date"].between('2020-12-31','2020-01-01')]
pricedate = data.sort_values(by = "Price Date")
pricedate
```

```python
# Now convrt this date wise record for further operation (It's an choice to convert it to the csv... But saving to the csv file is a good choice )
pricedate.to_csv('Agmarknet_Price.csv')
```
